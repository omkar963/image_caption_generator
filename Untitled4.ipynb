```python
CAPTION_FILE = "/kaggle/input/flickr8k/captions.txt"
IMAGE_DIR = "/kaggle/input/flickr8k/Images/"
```

```python
import os, string, pickle
import numpy as np
from tqdm import tqdm
from PIL import Image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
def load_captions(filename):
    with open(filename, 'r') as f:
        captions = f.readlines()
    mapping = {}
    for line in captions:
        line = line.strip()
        if len(line) < 1: continue
        img, caption = line.split(',', 1) if ',' in line else line.split('\t', 1)
        img = img.split('#')[0]
        caption = caption.lower().translate(str.maketrans('', '', string.punctuation))
        caption = 'startseq ' + caption + ' endseq'
        mapping.setdefault(img, []).append(caption)
    return mapping

captions_dict = load_captions(CAPTION_FILE)
print("Total images with captions:", len(captions_dict))
```

```python
from tensorflow.keras.applications.vgg16 import VGG16

model_vgg = VGG16(weights='imagenet')
model_vgg = Model(inputs=model_vgg.inputs, outputs=model_vgg.layers[-2].output)

def extract_features(directory):
    features = {}
    for img_name in tqdm(os.listdir(directory)):
        filename = os.path.join(directory, img_name)
        try:
            img = load_img(filename, target_size=(224,224))
            img = img_to_array(img)
            img = np.expand_dims(img, axis=0)
            img = preprocess_input(img)
            feature = model_vgg.predict(img, verbose=0)
            features[img_name] = feature
        except:
            continue
    return features

features = extract_features(IMAGE_DIR)
np.save('features.npy', features)
```

```python
all_captions = [c for caps in captions_dict.values() for c in caps]
tokenizer = Tokenizer()
tokenizer.fit_on_texts(all_captions)

vocab_size = len(tokenizer.word_index) + 1
max_length = max(len(c.split()) for c in all_captions)

with open('tokenizer.pkl','wb') as f:
    pickle.dump(tokenizer,f)

print("Vocabulary size:", vocab_size)
print("Max caption length:", max_length)
```

```python
import string

def load_captions(filename):
    with open(filename, 'r') as f:
        captions = f.readlines()
    mapping = {}
    for line in captions:
        line = line.strip()
        if len(line) < 1:
            continue
        # For Flickr8k dataset (captions.txt format: image_name, caption)
        img, caption = line.split(',', 1)
        caption = caption.lower().translate(str.maketrans('', '', string.punctuation))
        caption = 'startseq ' + caption + ' endseq'
        mapping.setdefault(img, []).append(caption)
    return mapping

CAPTION_FILE = "/kaggle/input/flickr8k/captions.txt"  # make sure path matches your dataset
captions_dict = load_captions(CAPTION_FILE)

print("Total images with captions:", len(captions_dict))


```

```python
import numpy as np

# Load pre-extracted image features
features = np.load('features.npy', allow_pickle=True).item()

print("Loaded features for", len(features), "images")
```

```python
def create_sequences(tokenizer, max_length, desc_list, photo, vocab_size):
    X1, X2, y = list(), list(), list()
    for desc in desc_list:
        seq = tokenizer.texts_to_sequences([desc])[0]
        for i in range(1, len(seq)):
            in_seq, out_seq = seq[:i], seq[i]
            in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
            out_seq = to_categorical([out_seq], num_classes=vocab_size)[0]
            X1.append(photo)
            X2.append(in_seq)
            y.append(out_seq)
    return np.array(X1), np.array(X2), np.array(y)

X1, X2, y = list(), list(), list()
for key, desc_list in tqdm(captions_dict.items()):
    photo = features.get(key)
    if photo is None: continue
    in_img, in_seq, out_word = create_sequences(tokenizer, max_length, desc_list, photo.flatten(), vocab_size)
    X1.append(in_img); X2.append(in_seq); y.append(out_word)

X1, X2, y = np.vstack(X1), np.vstack(X2), np.vstack(y)
print("Shapes:", X1.shape, X2.shape, y.shape)
```

```python
# Image feature branch
inputs1 = Input(shape=(4096,))
fe1 = Dropout(0.5)(inputs1)
fe2 = Dense(256, activation='relu')(fe1)

# Sequence branch
inputs2 = Input(shape=(max_length,))
se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
se2 = Dropout(0.5)(se1)
se3 = LSTM(256)(se2)

# Decoder (merge)
decoder1 = add([fe2, se3])
decoder2 = Dense(256, activation='relu')(decoder1)
outputs = Dense(vocab_size, activation='softmax')(decoder2)

model = Model(inputs=[inputs1, inputs2], outputs=outputs)
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.summary()
```

```python
history = model.fit([X1, X2], y, epochs=10, batch_size=64)
model.save('caption_model.h5')
```

```python
def word_for_id(integer, tokenizer):
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None

def generate_caption(model, tokenizer, photo, max_length):
    in_text = 'startseq'
    for _ in range(max_length):
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = model.predict([photo, sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = word_for_id(yhat, tokenizer)
        if word is None: break
        in_text += ' ' + word
        if word == 'endseq': break
    return in_text
```

```python
sample_img = os.path.join(IMAGE_DIR, os.listdir(IMAGE_DIR)[0])
img = load_img(sample_img, target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
photo = model_vgg.predict(img_array, verbose=0)

caption = generate_caption(model, tokenizer, photo, max_length)
print("Generated Caption:", caption)

plt.imshow(img)
plt.axis('off')
plt.title(caption)
plt.show()
```
